```python
import pandas as pd
```

```python
distances_FU = pd.read_csv('distances.csv')
relation_FU_price = pd.read_csv('FU-price.csv')['Price']
```

```python
def FU_to_price(elem) :
    if isinstance(elem, str) or pd.isna(elem):
        print(elem)
        return elem
    else :
        if elem > 200 :
            return 29.4
        return relation_FU_price[int(elem)]

def to_int(mystr) :
    return int(mystr) if (not pd.isna(mystr)) and (isinstance(mystr, float) or mystr.isnumeric()) else mystr
```

```python
distances_FU = distances_FU.map(to_int)
distances_FU
```

```python
price_matrix = distances_FU.map(FU_to_price)
```

```python
price_matrix
```

```python
price_matrix.to_csv('prices.csv')
```
